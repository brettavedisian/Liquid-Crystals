```python
%matplotlib inline
from sympy import *
from sympy import besselk
import numpy as np
from scipy.special import kn, jv
from scipy.integrate import quad
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
init_printing(use_unicode=True)
```

```python
x, y, r, R, theta_s, phi = symbols('x y r R theta_s phi')
```

```python
theta_s = 0.2
R = 1
```

```python
def nx(x, y, R, theta_s):
    return -theta_s*(R*y/(x**2 + y**2))

def ny(x, y, R, theta_s):
    return theta_s*(R*x/(x**2 + y**2))
```

```python
nx(x, y, R, theta_s), ny(x, y, R, theta_s)
```

```python
Y, X = np.mgrid[-3:3:15j, -3:3:15j]
NX = -theta_s*(R*Y/(X**2 + Y**2))
NY = theta_s*(R*X/(X**2 + Y**2))
U = np.sqrt(NX**2 + NY**2)
```

```python
plot1 = plt.figure(figsize=(11,8))
plt.quiver(X, Y, NX/U, NY/U, U, cmap=cm.copper, headlength=3)

plt.colorbar()

plt.title('Vector Field')
plt.show(plot1)
```
