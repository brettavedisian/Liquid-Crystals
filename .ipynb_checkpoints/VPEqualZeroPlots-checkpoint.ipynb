```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import csv
import time
```

```python
def B_a(a,gamma,beta):
    return a**3/beta**2+a**2*(3-1/beta**2)+a*(3*beta**2-3)+beta**4-2*beta**2+gamma

def Q_a(a,gamma,beta,rho):
    return (-(a/beta)**4*3
           +a**3*(-12/beta**2+2/beta**4)
           +a**2*((3*gamma+7)/beta**2-18)
           +a*(6*gamma-12*beta**2+8-2*gamma/beta**2)
           -3*beta**4+3*beta**2*gamma+3*beta**2-3*gamma+rho)

def R_a(a,gamma,beta,rho):
    return (a**6/beta**4
           +a**5*(6/beta**2+14/(3*beta**4))
           +a**4*(15+44/(3*beta**2)-31/(9*beta**4))
           +a**3*(20*beta**2+12-14*gamma/(3*beta**2)-86/(3*beta)**2)
           +a**2*(15*beta**4-(4*beta)**2/3-22*gamma/3-43/9+22*gamma/(3*beta)**2)
           +a*(6*beta**6-34*beta**4/3-2*beta**2*gamma/3+(4*beta)**2/3+2*gamma/3-20*rho/9)
           +beta**8-4*beta**6+2*beta**4*gamma+4*beta**4-4*beta**2*gamma+gamma**2)

# def B_crit(a,r_p,u_p,u_c,e,gamma):
#     return (u_p*(a/r_p)**3
#             +a**2*(3*u_p*gamma**4-2*e*(gamma*r_p)**2)/(gamma*r_p)**3
#             +a*(3*u_p*gamma**5-6*e*gamma**3*r_p**2)/(gamma*r_p)**3
#             +(gamma**6*u_p-4*e*gamma**4*r_p**2+u_c*gamma**2*r_p**4)/(gamma*r_p)**3)

# def Q_crit(a,r_p,u_p,u_c,v_c,e,gamma):
#     return (a**4*(-6*e*u_p*gamma**4*r_p)/(gamma*r_p)**5
#             +a**3*(-24*e*u_p*gamma**5*r_p+8*e**2*(gamma*r_p)**3)/(gamma*r_p)**5
#             +a**2*(-36*e*u_p*gamma**6*r_p+gamma**4*r_p**3*(28*e**2+3*u_c*u_p))/(gamma*r_p)**5
#             +a*(-24*e*u_p*gamma**7*r_p+2*gamma**5*r_p**3*(16*e**2+3*u_c*u_p)-4*e*u_c*r_p**5*gamma**3)/(gamma*r_p)**5
#             +(-6*e*u_p*gamma**8*r_p+3*gamma**6*r_p**3*(4*e**2+u_c*u_p)-6*e*u_c*r_p**5*gamma**4+v_c*r_p**6*gamma**4)/(gamma*r_p)**5)

# def R_crit(a,gamma,beta,rho):
#     return 9*B_a(a,gamma,beta)**2-20*a*Q_a(a,gamma,beta,rho)

def f(a,gamma,beta,rho,E,c):
    return (a*c**2)/2+(B_a(a,gamma,beta)*c**4)/4+(Q_a(a,gamma,beta,rho)*c**6)/6-E*c

def wupper(a,gamma,beta,rho):
    return ((-3*B_a(a,gamma,beta)+np.sqrt(R_a(a,gamma,beta,rho)))/(10*Q_a(a,gamma,beta,rho)))

def wlower(a,gamma,beta,rho):
    return ((-3*B_a(a,gamma,beta)-np.sqrt(R_a(a,gamma,beta,rho)))/(10*Q_a(a,gamma,beta,rho)))

def cupper(a,gamma,beta,rho):
    return np.sqrt(wupper(a,gamma,beta,rho))

def clower(a,gamma,beta,rho):
    return np.sqrt(wlower(a,gamma,beta,rho))

def ELOWER(a,gamma,beta,rho):
    return (a*cupper(a,gamma,beta,rho)+B_a(a,gamma,beta)*cupper(a,gamma,beta,rho)**3
            +Q_a(a,gamma,beta,rho)*cupper(a,gamma,beta,rho)**5)

def EUPPER(a,gamma,beta,rho):
    return (a*clower(a,gamma,beta,rho)+B_a(a,gamma,beta)*clower(a,gamma,beta,rho)**3
            +Q_a(a,gamma,beta,rho)*clower(a,gamma,beta,rho)**5)

def Elower(a,gamma,beta,rho):
    return (np.sqrt(10*(-3*B_a(a,gamma,beta)+np.sqrt(9*B_a(a,gamma,beta)**2
            -20*a*Q_a(a,gamma,beta,rho)))/Q_a(a,gamma,beta,rho))
            *(20*a*Q_a(a,gamma,beta,rho)-3*B_a(a,gamma,beta)**2
            +B_a(a,gamma,beta)*np.sqrt(9*B_a(a,gamma,beta)**2
            -20*a*Q_a(a,gamma,beta,rho)))/(250*Q_a(a,gamma,beta,rho)))

def Eupper(a,gamma,beta,rho):
    return (np.sqrt(10*(-3*B_a(a,gamma,beta)-np.sqrt(9*B_a(a,gamma,beta)**2
            -20*a*Q_a(a,gamma,beta,rho)))/Q_a(a,gamma,beta,rho))
            *(20*a*Q_a(a,gamma,beta,rho)-3*B_a(a,gamma,beta)**2
            -B_a(a,gamma,beta)*np.sqrt(9*B_a(a,gamma,beta)**2
            -20*a*Q_a(a,gamma,beta,rho)))/(250*Q_a(a,gamma,beta,rho)))
```

```python
start = time.time()

beta_val, gamma_val, rho_val = [], [], []

for i in np.linspace(-1,1,21): #beta
    for j in np.linspace(-1,1,21): #gamma
        for k in np.linspace(-1,1,21): #rho
            al = np.linspace(0,20,2001)
            if all(Q_a(al,j,i,k) > 0) and (min(B_a(al,j,i)) < 0):
                for n in np.linspace(0,4,41):
                    if (B_a(n,j,i) < 0) and (Q_a(n,j,i,k) > 0) and (R_a(n,j,i,k) > 0):
                        
                        beta_val.append(i)
                        gamma_val.append(j)
                        rho_val.append(k)
                        
vals = []
coeffs = ['beta','gamma','rho']
for i in range(len(beta_val)):
    vals.append(dict(zip(coeffs,[beta_val[i], gamma_val[i], rho_val[i]])))
    
def WriteDictToCSV(csv_file,csv_columns,dict_data):
    try:
        with open(csv_file, 'w+') as csvfile:
            writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
            writer.writeheader()
            for data in dict_data:
                writer.writerow(data)
    except IOError as err:
        errno,strerror = err.args
        print("I/O error({0}): {1}".format(errno, strerror))    
    return

WriteDictToCSV("np_vp_params.csv",coeffs,vals)

end = time.time()
print(end-start)
```
