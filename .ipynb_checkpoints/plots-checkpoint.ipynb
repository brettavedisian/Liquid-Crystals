```python
# %matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.axis as ax
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LightSource
from matplotlib.tri import Triangulation, UniformTriRefiner, CubicTriInterpolator
import numpy as np
import numpy.ma as ma
from scipy.special import kn, k1, k0, jn
from scipy.interpolate import griddata
from scipy.integrate import quad
from matplotlib import cm
```

```python
def r(x,y):
    return np.sqrt(x**2 + y**2)

def Nx(x,y,thetaS):
    return -y*thetaS/r(x,y)**2

def Ny(x,y,thetaS):
    return x*thetaS/r(x,y)**2

def n_perp(x, y, thetaS):
    return np.sqrt(Nx(x, y, thetaS)**2 + Ny(x, y, thetaS)**2)

def Nx_E(x, y, d, r1, thetaS):
    return -(thetaS*y*k1(r(x,y)*d**-1)/(r(x,y)*k1(r1*d**-1)))

def Ny_E(x, y, d, r1, thetaS):
    return (thetaS*x*k1(r(x,y)*d**-1)/(r(x,y)*k1(r1*d**-1)))

def nE_perp(x, y, d, r1, thetaS):
    return np.sqrt(Nx_E(x, y, d, r1, thetaS)**2 + Ny_E(x, y, d, r1, thetaS)**2)

def simple_f(x, y, d, r1, thetaS):
    return (thetaS**2*(k0(r(x,y)*d**-1)**2 + k1(r(x,y)*d**-1)**2))/(2*(d*k1(r1*d**-1))**2)

# def simple_f2(r, d, r1, thetaS):
#     return (thetaS**2*(k0(r/d)**2 + k1(r/d)**2))/(2*(d*k1(r1/d))**2)
```

```python
xaxis = np.zeros(1000)
yaxis = np.zeros(1000)
x = np.linspace(1,10, 1000)
y = np.linspace(1,10, 1000)

r1 = 1
dlist = [1,2,3]

colors = ['r','g','b']

dlabels = ['d = 1','d = 2','d = 3']
```

# 2 Cyl E-Field

###Attempt at 3D surface plot

```python
a = 2
d = 1
R_L = 1
R_R = 1
theta_1 = 0.2
theta_2 = 0.2
n_angles = 100
n_radii = 40
```

```python
# radii = np.linspace(1,10,n_radii)
# angles = np.linspace(0,2*np.pi,n_angles,endpoint=False)
```

```python
# angles = np.repeat(angles[...,np.newaxis],n_radii,axis=1)
```

```python
# xX = np.append(0,(radii*np.cos(angles)).flatten())
# yY = np.append(0,(radii*np.sin(angles)).flatten())
```

```python
def surf_eqn(x,y,a,d,theta_1, theta_2):
    return ((theta_1*kn(0,np.sqrt(a**2-2*a*x+x**2+y**2)/d) + theta_2*kn(0,np.sqrt(a**2+2*a*x+x**2+y**2)/d))/(d*k1(1/d)))**2/2
```

```python
def n_eqn(x,y,a,d,theta_1, theta_2):
    return (((y**2*(theta_2*np.sqrt(y**2+(x-a)**2)*k1(np.sqrt(y**2+(x+a)**2)/d)
           + theta_1*np.sqrt(y**2+(x+a)**2)*k1(np.sqrt(y**2+(x-a)**2)/d))**2)
           - (y**2+(x-a)**2)*(y**2+(x+a)**2)*k1(1/d)**2
           + (-(x+a)*theta_2*np.sqrt(y**2+(x-a)**2)*k1(np.sqrt(y**2+(x+a)**2)/d)
           + (a-x)*theta_1*np.sqrt(y**2+(x+a)**2)*k1(np.sqrt(y**2+(x-a)**2)/d))**2)
            /(2*d**2*(y**2+(x-a)**2)*(y**2+(x+a)**2)*k1(1/d)**2))
```

```python
xX, yY = np.meshgrid(np.linspace(-5,5,100),np.linspace(-5,5,100))
```

```python
xX = xX.flatten()
yY = yY.flatten()
```

```python
xX.shape
```

```python
triang = Triangulation(xX,yY)
xmid = xX[triang.triangles].mean(axis=1)
ymid = yY[triang.triangles].mean(axis=1)
mask1 = np.where(np.logical_and(((xmid-a)*(xmid-a) + ymid*ymid), ((xmid+a)*(xmid+a) + ymid*ymid)) < 1, 1, 0)
triang.set_mask(mask1)
```

```python
ZZ = surf_eqn(triang.x,triang.y,a,d,theta_1,theta_2) + n_eqn(triang.x,triang.y,a,d,theta_1,theta_2)
```

```python
ZZ
```

```python
fig = plt.figure(figsize=(11,8))
ax = fig.gca(projection='3d')
ax.view_init(30,90)
ax.set_aspect('equal')
# ax.dist=5
# ZZ[ZZ>4] = np.nan
surf = ax.plot_trisurf(triang, ZZ, cmap=cm.jet,linewidth=0)
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
ax.set_zlabel('f', fontsize=14)
ax.set_title('Electric Field Free Energy Density', fontsize=18)
fig.colorbar(surf,shrink=0.7)
ax.set_zlim3d(0,5)
ax.set_xlim3d(-5,5)
ax.set_ylim3d(-5,5)
fig.tight_layout();
```

# Single Cyl

```python
radii_single = np.linspace(1,10,n_radii)
angles_single = np.linspace(0,2*np.pi,n_angles,endpoint=False)
```

```python
angles_single = np.repeat(angles_single[...,np.newaxis],n_radii,axis=1)
```

```python
x_single = (radii_single*np.cos(angles_single)).flatten()
y_single = (radii_single*np.sin(angles_single)).flatten()
```

```python
tri = Triangulation(x_single,y_single)
xmid = x_single[tri.triangles].mean(axis=1)
ymid = y_single[tri.triangles].mean(axis=1)
mask = np.where(xmid*xmid + ymid*ymid < 1, 1, 0)
tri.set_mask(mask)
```

```python
def single(x, y, theta_s):
    return 8*np.tan(theta_s/2)**6/(x**2+y**2+np.tan(theta_s/2)**2)**4
```

```python
work = n_perp(tri.x,tri.y,0.2)
```

```python
xx = np.append(-radii_single[::-1],radii_single)
yy = xx.copy()
```

```python
pos = np.where(abs(xx)==1)[0]
```

```python
xx[pos] = np.nan
yy[pos] = np.nan
```

```python
n_perp(xx,yy,0.2)
```

```python
plt.plot(xx,single(xx,yy,0.2))
```
