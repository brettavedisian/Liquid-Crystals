```python
%matplotlib inline
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
init_printing(use_unicode=True)
```

```python
r, T, u, v, c = symbols('r T u v c')
```

```python
def f_sympy(T, u, c):
    return (r(T)*c**2)/2 + (u*c**4)/4

def f_numpy(r, u, c):
    return (r*c**2)/2 + (u*c**4)/4
```

```python
f = f_sympy(T, u, c)
```

```python
f.diff(c)
```

```python
n = 2
```

```python
T = np.linspace(-n,n,51)
```

```python
fig1 = plt.figure(figsize=(11,8))
ax1 = fig1.gca()
plt.plot(T, f_numpy(1, 1, T))
plt.xlabel('c', fontsize=14)
plt.ylabel('f', rotation='horizontal',verticalalignment='center',  fontsize=14)
ax1.yaxis.set_label_coords(0.42,1)
ax1.xaxis.set_label_coords(1.05,0.02)
ax1.spines['left'].set_position('zero')
ax1.spines['right'].set_color('none')
ax1.spines['top'].set_color('none')
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
yticks1 = ax1.yaxis.get_major_ticks()
yticks1[0].label1.set_visible(False);
```

```python
def c(r,u):
    a = []
    for i in r:
        if i < 0:
            a.append(np.sqrt(-i/u))
        else:
            a.append(0)
    return np.array(a)
```

```python
x = np.linspace(0,2,100)
```

```python
plt.figure(figsize=(11,8))
plt.plot(x, c(x-0.5,1),label='+')
plt.plot(x, -c(x-0.5,1),label='-')
plt.xlabel('T',fontsize=14)
plt.ylabel('c',fontsize=14,rotation='horizontal')
plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')
plt.gca().spines['bottom'].set_position('zero')
plt.gca().xaxis.set_ticks_position('bottom')
plt.gca().yaxis.set_ticks_position('left')
plt.legend(fontsize=18)
plt.gca().xaxis.set_ticklabels(['', '$T_{C}$','','',''],fontsize=14)
plt.gca().xaxis.set_label_coords(1.02,0.515);
```
