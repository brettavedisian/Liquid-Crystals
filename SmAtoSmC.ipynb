```python
%matplotlib inline
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
init_printing(use_unicode=True)
```

```python
r, T, u, v, c = symbols('r T u v c')
```

#u > 0

Solution for $c_{min}$:

$c_{min} = \pm\sqrt{\dfrac{-r}{u}}$

```python
def f1_sympy(T, u, c):
    return (r(T)*c**2)/2 + (u*c**4)/4

def f1_numpy(r, u, c):
    return (r*c**2)/2 + (u*c**4)/4
```

```python
f1 = f1_sympy(T, u, c)
```

```python
f1.diff(c)
```

###c vs T

```python
def c1(r,u):
    a = []
    for i in r:
        if i < 0:
            a.append(np.sqrt(-i/u))
        else:
            a.append(0)
    return np.array(a)
```

#u < 0

From Wolfram Alpha (two solutions for $c_{min}$:

$c_{min} = \pm\sqrt{\dfrac{|u| - \sqrt{|u|^{2} - 4r|v|}}{2|v|}}$

$c_{min} = \pm\sqrt{\dfrac{|u| + \sqrt{|u|^{2} - 4r|v|}}{2|v|}}$

```python
def f2_sympy(T, u, v, c):
    return (r(T)*c**2)/2 - (u*c**4)/4 + (v*c**6)/6

def f2_numpy(r, u, v, c):
    return (r*c**2)/2 - (abs(u)*c**4)/4 + (abs(v)*c**6)/6
```

```python
f2 = f2_sympy(T, u, v, c)
```

```python
f2.diff(c)
```

```python
def c2(r, u, v):
    a = []
    for i in r:
        if i < 0:
            a.append(np.sqrt((abs(u)+np.sqrt(abs(u)**2-4*i*abs(v)))/(2*abs(v))))
        else:
            a.append(0)
    return np.array(a)
```
